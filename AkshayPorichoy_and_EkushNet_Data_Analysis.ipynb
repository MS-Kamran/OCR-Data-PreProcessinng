# **Data Process**

# import

```python
# prompt: load a csv file using pandas for data analysis

import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/mskamran/Documents/Data Science/final letters process/CSV files/Akshar_Porichoy_and_Ekuh_Net.csv')

# Print the DataFrame
print(df)
```

# Split in column

```python
# Splitting the 'Image Name' column based on '_' and '.'
df_split = df['Image Name'].str.split('_|\.', expand=True)

# Renaming the columns for clarity
df_split.columns = ['Division', 'School', 'Class', 'Gender', 'Age', 'Page1', 'Code2', 'Letters', 'Filename', 'Extension']

# Concatenating the split DataFrame with the original DataFrame
df = pd.concat([df, df_split], axis=1)

# Dropping the original 'Image Name' column if necessary
# df = df.drop(columns=['Image Name'])

# Displaying the resulting DataFrame
print(df)
```

```python
# Set the maximum column width to display full contents
pd.set_option('display.max_colwidth', None)

# Display the first 100 rows of the DataFrame
print(df.head(10).to_string(index=False))
```

# Delete Some Column

```python
df.drop(columns=['Image Name'], inplace=True)
df.drop(columns=['Code2'], inplace=True)
df.drop(columns=['Filename'], inplace=True)
df.drop(columns=['Extension'], inplace=True)

print(df.head(10).to_string(index=False))
```

# **Divisions**

```python
# prompt: Division column unique value and count

unique_divisions = df['Division'].unique()
division_counts = df['Division'].value_counts()

print("Unique Divisions:")
print(unique_divisions)

print("\nDivision Counts:")
print(division_counts)


# prompt: Division column unique value and count with graph (various color and values are on top of the graphs) Use Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

# Get unique values and counts for the 'Division' column
unique_divisions = df['Division'].unique()
division_counts = df['Division'].value_counts()

# Create a bar chart with various colors and values on top of the bars
plt.figure(figsize=(12, 6))
sns.barplot(x=unique_divisions, y=division_counts, palette='Set2')

# Add values on top of each bar
for i, count in enumerate(division_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Set title and axis labels
plt.title('Division Counts', fontsize=16)
plt.xlabel('Division', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# **Class**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the desired order of classes
desired_order = ['C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9', 'C10', 'C11', 'C12', 'C11C12', 'UNI']

# Get unique values and counts for the 'Class' column
unique_Class = df['Class'].unique()
Class_counts = df['Class'].value_counts()

# Sort Class_counts based on the desired order
Class_counts_sorted = Class_counts.reindex(desired_order)

# Print the results
print("Unique Classes:")
print(unique_Class)

print("\nClass Counts:")
print(Class_counts_sorted)

# Plotting the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=Class_counts_sorted.index, y=Class_counts_sorted.values, palette="viridis")
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Class Counts (Custom Order)')
plt.xticks(rotation=45)

# Displaying values on top of the bars
for i, count in enumerate(Class_counts_sorted.values):
    plt.text(i, count + 0.1, str(count), ha='center')

plt.show()
```

# **Gender**

```python
import matplotlib.pyplot as plt

# Preprocess the 'Gender' column to replace 'Others' with 'Null'
df['Gender'] = df['Gender'].apply(lambda x: x if x in ['F', 'M'] else 'Null')

# Get unique values and counts for the 'Gender' column
unique_genders = df['Gender'].unique()
gender_counts = df['Gender'].value_counts()

# Print the results
print("Unique Genders:")
print(unique_genders)

print("\nGender Counts:")
print(gender_counts)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Gender Distribution')
plt.show()
```
