# **Data Process**

# import

```python
# prompt: load a csv file using pandas for data analysis

import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('/Users/mskamran/Documents/Data Science/All Types Of Datas/Ekush New Data from Online/Data process/EkushNet_image_names.csv')

# Print the DataFrame
print(df)
```

# Split in column

```python
# Splitting the 'Image Name' column based on '_' and '.'
df_split = df['Image Name'].str.split('_|\.', expand=True)

# Renaming the columns for clarity
df_split.columns = ['Division', 'District', 'Class', 'Gender', 'Age', 'Code1', 'Code2', 'Letters', 'Filename', 'Extension']

# Concatenating the split DataFrame with the original DataFrame
df = pd.concat([df, df_split], axis=1)

# Dropping the original 'Image Name' column if necessary
# df = df.drop(columns=['Image Name'])

# Displaying the resulting DataFrame
print(df)
```

```python
# Set the maximum column width to display full contents
pd.set_option('display.max_colwidth', None)

# Display the first 100 rows of the DataFrame
print(df.head(10).to_string(index=False))
```

# Delete Some Column

```python
df.drop(columns=['Image Name'], inplace=True)
df.drop(columns=['Filename'], inplace=True)
df.drop(columns=['Extension'], inplace=True)

print(df.head(10).to_string(index=False))
```

# **Divisions**

```python
# prompt: Division column unique value and count

unique_divisions = df['Division'].unique()
division_counts = df['Division'].value_counts()

print("Unique Divisions:")
print(unique_divisions)

print("\nDivision Counts:")
print(division_counts)


# prompt: Division column unique value and count with graph (various color and values are on top of the graphs) Use Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

# Get unique values and counts for the 'Division' column
unique_divisions = df['Division'].unique()
division_counts = df['Division'].value_counts()

# Create a bar chart with various colors and values on top of the bars
plt.figure(figsize=(12, 6))
sns.barplot(x=unique_divisions, y=division_counts, palette='Set2')

# Add values on top of each bar
for i, count in enumerate(division_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

# Set title and axis labels
plt.title('Division Counts', fontsize=16)
plt.xlabel('Division', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

# **Category**

```python
import matplotlib.pyplot as plt

# Define a dictionary mapping old values to new labels
category_mapping = {
    '0': 'Primary',
    '1': 'High School',
    'C11C12': 'College',
    'UNI': 'University',
    'OTH': 'Other Occupation',
    'NULL': 'Null'
}

# Map the values in the 'Category' column to their corresponding labels
df['Category'] = df['Category'].map(category_mapping)

# Get unique values and counts for the 'Category' column
unique_categories = df['Category'].unique()
category_counts = df['Category'].value_counts()

# Print the results
print("Unique Categories:")
print(unique_categories)

print("\nCategory Counts:")
print(category_counts)

# Plotting the category counts
plt.figure(figsize=(10, 6))
barplot = category_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Counts')

# Add count numbers on top of each bar
for index, value in enumerate(category_counts):
    plt.text(index, value + 5, str(value), ha='center', va='bottom')

plt.show()
```

# **Gender**

```python
import matplotlib.pyplot as plt

# Preprocess the 'Gender' column to replace 'Others' with 'Null'
df['Gender'] = df['Gender'].apply(lambda x: x if x in ['F', 'M'] else 'Null')

# Get unique values and counts for the 'Gender' column
unique_genders = df['Gender'].unique()
gender_counts = df['Gender'].value_counts()

# Print the results
print("Unique Genders:")
print(unique_genders)

print("\nGender Counts:")
print(gender_counts)

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Gender Distribution')
plt.show()
```

# **AGE**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Preprocess the 'Age' column to filter valid age values
df['Age'] = df['Age'].apply(lambda x: int(x[1:]) if (x.startswith('A') and x[1:].isdigit()) else None)

# Remove NULL values
df = df.dropna()

# Get unique values and counts for the 'Age' column
age_counts = df['Age'].value_counts().sort_index()

# Plotting the histogram
plt.figure(figsize=(8, 8))
plt.hist(df['Age'], bins=range(int(df['Age'].min()), int(df['Age'].max()) + 2), color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution')

# Adding count values on top of bars
for i, count in enumerate(age_counts):
    plt.text(age_counts.index[i], count + 0.5, str(count), ha='center', va='bottom')

plt.grid(axis='y')
plt.show()
```
